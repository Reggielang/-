```python
import requests

res = requests.get('http://httpbin.org/get')
print(res.text)
```

```python
#带参数的请求
import requests
res = requests.get('http://httpbin.org/get?name=germey&age=22')
print(res.text)
```

```python
#把参数提前写好，再传入请求当中
import requests

data = {
    'name':'germey',
    'age':22
}

res = requests.get('http://httpbin.org/get',params=data)
print(res.text)
```

```python
#解析json
import requests
import json

res = requests.get('http://httpbin.org/get')
print(type(res.text))
print(res.json())
print(json.loads(res.text))
print(type(res.json()))
```

# 高级操作

```python
#文件上传
import requests

files = {'file':open('favicon.ico','rb')}
res = requests.post('http://httpbin.org/post',files=files)
print(res.text)
```

```python
#获取cookies
import requests

res = requests.get('http://www.baidu.com')
print(res.cookies)
for key,value in res.cookies.items():
    print(key + '=' + value)
```

```python
#会话维持
#其实这是两个不同的请求，完全独立的过程。用一个请求设置了cookies，但是用另一个发出请求，所以获取不到cookies信息
import requests

requests.get('http://httpbin.org/cookies/set/number/123456')
res = requests.get('http://httpbin.org/cookies')
print(res.text)
```

```python
#Session可以提供多次请求的对象
import requests

s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456')
res = s.get('http://httpbin.org/cookies')
print(res.text)
```

```python
#证书验证
import requests

res = requests.get('https://www.12306.cn')
print(res.status_code)
```

```python
#如果是爬取证书错误的网页时，加入后面的参数即可，同时如果想要避免有警告信息，可以导入urllib3来解决
import requests
from requests import urllib3
urllib3.disable_warnings()
res = requests.get('https://www.12306.cn',verify=False)
print(res.status_code)
```

```python
#代理设置
import requests

proxies = {
  "http": "http://127.0.0.1:9743",
  "https": "https://127.0.0.1:9743",
}

response = requests.get("https://www.taobao.com", proxies=proxies)
print(response.status_code)
```

```python
import requests

proxies = {
    "http": "http://user:password@127.0.0.1:9743/",
}
response = requests.get("https://www.taobao.com", proxies=proxies)
print(response.status_code)
```

```python
import requests

proxies = {
    'http': 'socks5://127.0.0.1:9742',
    'https': 'socks5://127.0.0.1:9742'
}
response = requests.get("https://www.taobao.com", proxies=proxies)
print(response.status_code)
```

```python
#超时设置
import requests
from requests.exceptions import ReadTimeout
try:
    res = requests.get('http://www.taobao.com',timeout = 0.1)
    print(res.status_code)
except ReadTimeout:
    print('Timeout')
```

```python
#认证设置
#需要登陆验证的网站
import requests
from requests.auth import HTTPBasicAuth

r = requests.get('http://120.27.34.24:9001', auth=HTTPBasicAuth('user', '123'))
print(r.status_code)
```

```python
#异常处理
import requests
from requests.exceptions import ReadTimeout,ConnectionError,RequestException
try:
    res = requests.get('http://httpbin.org/get',timeout = 0.5)
    print(res.text)
except ReadTimeout:
    print('Timeout')
except ConnectionError:
    print('Connection error')
except RequestException:
    print('Error')
```
