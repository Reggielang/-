```python
import requests
```

```python
res = requests.get('http://www.baidu.com')
#响应体
print(res.text)
```

```python
#响应头
print(res.headers)
```

```python
#响应码
print(res.status_code)
```

```python
#包含响应头的请求
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'}
res = requests.get('http://www.baidu.com',headers = headers)
print(res.status_code)
```

```python
#图片的抓取
res = requests.get('https://dss0.bdstatic.com/5aV1bjqh_Q23odCf/static/mancard/img/card_setts-c75091e958.png',headers = headers)
print(res.content)
```
